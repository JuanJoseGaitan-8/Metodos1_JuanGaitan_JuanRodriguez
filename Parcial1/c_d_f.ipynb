```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.01
xi, xf = 0.1, 1.1
Npoints = int((xf-xi)/h)
```

```python
x = np.linspace(xi,xf,Npoints)
```

```python
# Definición de función
def f(x):
    return (np.sqrt(np.tan(x)))
# Definición de la primera derivada exacta
def df1(x):
    return (1/(np.cos(x))**2)/(2* np.sqrt(np.tan(x)))
```

```python
Dy = df1(x)
```

```python
def DerivadaCentral(f,x,h):
    d = 0.
    if h!=0:
        d = ( f(x+h) - f(x-h) ) / ( 2*h )
    return d
```

```python
def DerivadaProgresiva(f,x,h):
    a = 0
    if h!=0:
        a = (1/(2*h)) * ( ( -3*(f(x)) )+ ( 4*f(x+h) ) - ( f(x+(2*h))) )
    return a
```

```python
Dcp = DerivadaProgresiva(f,x,h)

Dcy = DerivadaCentral(f,x,h)
```

```python
plt.plot(x,Dy,c='g',label='Derivada Exacta')
plt.plot(x,Dcy,c='b',label='Derivada Central')
plt.legend(loc="upper right") 
plt.show()
```

```python
plt.plot(x,Dy,c='g',label='Derivada Exacta')
plt.plot(x,Dcp,c='red',label='Derivada Progresiva')
plt.legend(loc="upper right") 
plt.show()
```

```python
# Derivada central

ErrorNodalCentral = np.abs(Dcy-Dy)
print('El error local para cada nodo usando la derivada central es:')
print(ErrorNodalCentral)
```

```python
plt.plot(x,ErrorNodalCentral, c='green')
plt.title("Estimación del error en cada punto nodal dentro del intervalo.", size=10, c='black')
plt.show()
```

```python
# Derivada progresiva

ErrorNodalProgresiva = np.abs(Dcp-Dy)
print('El error local para cada nodo usando la derivada central es:')
print(ErrorNodalProgresiva)
```

```python
plt.plot(x,ErrorNodalProgresiva, c='brown')
plt.title("Estimación del error en cada punto nodal dentro del intervalo.", size=10, c='black')
plt.show()
```
