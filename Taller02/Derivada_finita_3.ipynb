```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

## Punto 1.3

```python
h = 0.05
xi, xf = -10., 10.
Npoints = int((xf-xi)/h)
```

```python
x = np.linspace(xi,xf,Npoints)
```

```python
# Definición de función
def f(x):
    return 1 / (np.sqrt(1+np.exp(-x**2)))
# Definición de la (M) mascara
Mascara = np.array([1.0, 0, -1.0])
```

```python
def operador_derivada_central(f,x,h,mascara):
    y = np.zeros(len(x))
    for i in range(1, len(x)-1):
        resultado = np.array([f(x[i-1]), f(x[i]), f(x[i+1])])
        y[i] = np.dot((resultado/2*h), Mascara)
    return y
```

```python
convolucion = operador_derivada_central(f,x,h,Mascara)

plt.plot(x, convolucion,  c='red', label='Operador Derivada Central')
plt.legend(loc="upper right") 
plt.title("Derivada punto 1.2 usando kernel", size=10, c='black')
plt.show()
```
