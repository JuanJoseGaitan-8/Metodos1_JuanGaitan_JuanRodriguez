```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

```

```python
xi, xf, N = -4,4,25
x = np.linspace(xi,xf,N)
y = x.copy()
r = 2
v = 2
h = 0.001
#X,Y = np.meshgrid(x,y)
```

```python
def Potencial(x,y):
    return (v*x-(((v*x)*(r**2)/((x))**2 + (y)**2) - ((v*x)-((v*x)*(r**2)/(( (x)**2)) + (y)**2  ))))
```

```python
def Dx(f,x,y,h):
    if np.sqrt(x**2+y**2)<= 2:
        return 0
    else:
        return (f(x+h,y)-f(x-h,y))/(2*h)
    
def Dy(f,x,y,h):
    if np.sqrt(x**2+y**2)<=2:
        return 0
    else:
        return -(f(x,y+h)-f(x,y-h))/(2*h)
```

```python
Ex = np.zeros((N,N))
Ey = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        Ex[i,j] = Dx(Potencial,x[i],y[j],h)
        Ey[i,j] = -(Dy(Potencial,x[i],y[j],h))


Ex = Ex.T
Ey = Ey.T
```

```python
fig1 = plt.figure(figsize=(5,5))
ax = fig1.add_subplot(1,1,1)

for i in range(1,len(x)):
    for j in range(1,len(y)):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j],color='b')
```
