## Punto 1.4

b) Implementar el cálculo de esta derivada a la función del punto 1.2


```python
h = 0.05
xi, xf = -10., 10.
Npoints = int((xf-xi)/h)
```

```python
x = np.linspace(xi,xf,Npoints)
```

```python
# Definición de la función
def f(x):
    return 1 / (np.sqrt(1+np.exp(-x**2)))
```

```python
# Definición de la (M) mascara
Mascara = np.array([1.0, -2.0, 1.0])
```

```python
def operador_segundaderivada_central(f,x,h,m):
    y = np.zeros(len(x))
    for i in range(1, len(x)-1):
        resultado = np.array([f(x[i-1]), f(x[i]), f(x[i+1])])
        y[i] = np.dot((resultado/2*h), m)
    return y
```

```python
convolucion2 = operador_segundaderivada_central(f,x,h,Mascara)

plt.plot(x, convolucion2,  c='blue', label='Operador Derivada Central')
plt.legend(loc="upper right") 
plt.title("Derivada df2 usando kernel", size=10, c='black')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def segunda(h):
  f = lambda x: 1/np.sqrt(1+np.exp(-x**2))
  X=np.arange(-10,10,h)
  Y=np.zeros(len(X))
  M=np.array([1,0,-2,0,1])
  for i in range(2,len(X)-2):
    F=np.array([f(X[i-2]),f(X[i-1]),f(X[i]),f(X[i+1]),f(X[i+2])])
    Y[i]=np.dot(F/(4*(h**2)),M)
  return Y

h=0.05
y=segunda(h)
x=np.arange(-10,10,h)
fig = plt.figure(figsize=(7,3))
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(x,y)
```
