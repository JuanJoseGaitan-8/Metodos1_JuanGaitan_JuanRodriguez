```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

## Punto 1.2

```python
h = 0.05
xi, xf = -10., 10.
Npoints = int((xf-xi)/h)
```

```python
x = np.linspace(xi,xf,Npoints)
```

```python
# Definición de función
def f(x):
    return 1 / (np.sqrt(1+np.exp(-x**2)))
# Definición de la primera derivada exacta
def df1(x):
    return (x*np.exp(-x**2))/(np.exp(-x**2)+1)**(3/2)
```

```python
Dy = df1(x)
```

```python
def DerivadaCentral(f,x,h):
    d = 0.
    if h!=0:
        d = ( f(x+h) - f(x-h) ) / ( 2*h )
    return d
```

a) En el intervalo $-10\leq x\leq 10$

```python
Dcy = DerivadaCentral(f,x,h)
```

```python
plt.plot(x,Dy,c='g',label='Derivada Exacta')
plt.plot(x,Dcy,c='b',label='Derivada Central')
plt.legend(loc="upper right") 
plt.show()
```

b) Para el intervalo anterior, estimar el error en cada punto nodal.

```python

ErrorNodal = np.abs(Dcy-Dy)
```

```python
print('El error local para cada nodo es:')
print(ErrorNodal)
```

```python
plt.plot(x,ErrorNodal, c='red')
plt.title("Estimación del error en cada punto nodal dentro del intervalo.", size=10, c='black')
plt.show()
```
