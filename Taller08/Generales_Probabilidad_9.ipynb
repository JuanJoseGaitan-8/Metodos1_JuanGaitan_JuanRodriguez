```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python

p2 = np.linspace(0.1,0.5,100)
p1 = np.linspace(0.1,0.9,100) 

P1,P2 = np.meshgrid(p1,p2)

Z = (P1*P2+2*(1-P1)*P2+2*P1*(1-P2)+(1-P1)*(1-P2))/4
```

```python

figura = plt.figure( figsize=( 8 , 8 ) )
ax1 = figura.add_subplot( 1 , 1 , 1 , projection='3d' )    
surf = ax1.plot_surface( P1 , P2 , Z , cmap=cm.coolwarm , antialiased=True )
figura.colorbar(surf, shrink=0.6)   
ax1.set_xlabel('P1')

figura.suptitle( 'Probabilitys surface' )
ax1.set_ylabel('P2')
ax1.view_init(45,85)  
```
