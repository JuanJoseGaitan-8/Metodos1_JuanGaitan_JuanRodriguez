```python

import matplotlib.pyplot as plt
import numpy as np
```

```python

def productoria( rango ):
    
    prod = 1
    
    for i in range( rango ):
        prod *= ( 365 - i ) / 365
        
    return(prod)    
```

```python

lista = [] 
n = []

for i in range(80):
    lista.append(productoria(i))
    n.append(i)
    
plt.plot( n , lista , c='r',linestyle='-')
plt.xlabel( 'Número de cumpleaños diferentes', c='k' )
plt.ylabel( 'Probabilidad de exito', c='k' )

plt.show()
```
