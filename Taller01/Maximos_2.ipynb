```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("./Maximos_1.txt", header=None, sep="\s+" )

x = datos[0]
y = datos[1]

plt.plot( x , y , c = "green")
plt.show()
```

```python
def maximos_locales( datos ):
    
    p_criticos = []
    
    for i in range( 1 , max( datos.index ) ):
        if y[ i ] > y[ i+1 ] and y[ i ] > y[ i-1 ]:
            p_criticos.append( i )
            
    return p_criticos
```

```python
max_local = maximos_locales( datos )
```

```python
plt.plot( x , y , c = "green" )

plt.scatter( x[ max_local ], y[ max_local ] , c = "black" , marker = "o", s = 15 )

plt.show()
```
