```python
#5.1 Números primos
def numerosprimos():
    numeros = 2
    yield numeros
    
    while True:
        a = numeros
        while True:
            a += 1
            cont = 1
            numero_divisores = 0
            while cont <= a:
                if a % cont == 0:
                    numero_divisores +=1
                if numero_divisores > 2:
                    break
                cont += 1
            if numero_divisores ==2:
                yield a
b = numerosprimos()
c = numerosprimos()
#Punto 1. Calcular 1000 numeros primos
darnumero = [next(b) for _ in range(1000)] #Calcular los primeros 1000 números primos.

#Punto 2. Imprimir los primeros 10 números primos.
darnumerob = [next(c) for _ in range(10)]
print("Los primeros 10 numeros primos son:" , darnumerob) #Imprimir los primeros 10 números primos
            
    
    
    
```

```python
# 5.3 Figuras de Lissajous.
# Punto a. Con desfase igual a 0
from numpy import sin,linspace,pi
from pylab import plot, subplot, show

valores_x = [1,2,3,4,5]
valores_y = [1,3,5,7,9]
t = np.linspace(-pi,pi,300)
valores_des = 0
for i in range (0,5):
    x = sin(valores_x[i]*t)
    y = sin(valores_y[i]*t + valores_des)
    subplot(3,2,i+1)
    plot(x,y)
    plt.axis("off")

show()
```
