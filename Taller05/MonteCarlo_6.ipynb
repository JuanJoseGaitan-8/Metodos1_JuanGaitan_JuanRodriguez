```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from mpl_toolkits.mplot3d import axes3d
```

```python
def CreadorCirculo( N,  R ):
    
    X = np.zeros(N)
    Y = np.zeros_like(X)
    Z = np.zeros_like(X)
    
    
    for i in tqdm(range(N)):
        
        u = np.random.rand()
        r = R*u**(1/3)
        
        phi = np.random.uniform(0,np.pi)
        theta = np.random.uniform(0,2*np.pi)
      
        
        X[i] = r*np.sin(phi)*np.cos(theta)
        Y[i] = r*np.sin(phi)*np.sin(theta)
        Z[i] = r*np.cos(phi)
    
    
    return X,Y,Z
```

```python
X,Y,Z = CreadorCirculo( 10000, 1 )
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection = '3d')

#configurar límites de los ejes
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)

ax.view_init(10, 60)

ax.scatter(X,Y,Z,color='b')
```

```python
def f(x,y,z):
    return np.exp(np.sqrt((x**2)+(y**2)+(z**2)))
```

```python
N = int(1e6)
X,Y,Z = CreadorCirculo(N,1)
Sample = np.zeros(N)

for i in range(N):
    x = X[i]
    y = Y[i]
    z = Z[i]
    
    Sample[i] = f(x,y,z)
    
```

```python
Integral = (4/3)*np.pi * np.average( Sample )
```
