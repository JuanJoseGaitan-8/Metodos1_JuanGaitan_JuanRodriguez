```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
valores = 30
N = 10**4
K = np.linspace(0,valores,valores)
matriz = np.zeros(valores)
x = np.random.uniform(0,1,N)


```

```python
for i in range(valores):
    sumatoria_total = 0
    for j in range(0, N-1000):
        sumatoria = x[j]*x[j+i]
        sumatoria_total += sumatoria
    matriz[0] = sumatoria_total/N
    matriz[i] = sumatoria_total/N
         
```

```python
print(matriz)
```

```python

fig = plt.figure(figsize=(8, 6))
plt.plot(K, matriz)
plt.title("Generador Numpy",fontsize=25)
plt.xlabel("K-enesimo vecino",fontsize=18)
plt.ylabel("C(k)",fontsize=18)
plt.xlim(0,30)
plt.ylim(0.20,0.30)
plt.show()
```
